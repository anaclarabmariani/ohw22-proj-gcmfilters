# Make a movie

### Open data

```python
from intake import open_catalog
import xarray as xr
import matplotlib.pyplot as plt

cat = open_catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean/CESM_POP.yaml")
list(cat)
```

```python
ds = cat['CESM_POP_hires_RCP8_5'].to_dask()
```

```python
ssh = ds.SSH_2
ssh = ssh.chunk({'nlat': len(ds.nlat), 'nlon': len(ds.nlon)})  # 1 chunk in nlon and nlat direction
ssh
```

**ssh is a dask array wrapped in xarray!**

### Next: make a movie with global SSH

```python
import xmovie
from xmovie import Movie
# here is the full documentation: https://xmovie.readthedocs.io/en/latest/index.html
```

```python
mov = Movie(ssh.isel(time=slice(None, 10)))  # only make a movie from the first 10 time slices; this happens all "lazily" because we use dask input data
```

#### Let's get a view of the first time slice before we go for the full movie!

```python
mov.preview(0)  
plt.savefig("movie_preview.png")
```

#### The following will make one figure for each time slice and at the end concatenate them together to a full movie

```python
mov.save('movie.mp4')  
```
